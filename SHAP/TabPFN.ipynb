```python
import pandas as pd
import numpy as np
from tabpfn import TabPFNClassifier
import shap
import matplotlib.pyplot as plt
```

```python
seed = 873465341
```

```python
data = pd.read_csv('data_SHAP.csv')
X = data.iloc[:,2:]
y = data.iloc[:,1]
```

```python
model = TabPFNClassifier(random_state=seed)
model.fit(X,y)
```

```python
def func(X):
  return model.predict_proba(X)[:,1]
```

```python
explainer = shap.KernelExplainer(func,shap.sample(X,nsamples=20,random_state=seed))
shap_values = explainer.shap_values(X)
```

```python
pd.DataFrame(shap_values).to_csv('SHAP_values.csv',index=False)
```
